```python
import pymongo
from pymongo import MongoClient
```

```python
client = MongoClient("localhost", 27017)
```

```python
db = client.netflix
colecao = db.users
```

```python
# Recupere os documentos que contêm informações sobre os títulos assistidos
documentos = colecao.find()
```

```python
títulos_contagem = {}
```

```python
for documento in documentos:
    título = documento["Title"]
    if título in títulos_contagem:
        títulos_contagem[título] += 1
    else:
        títulos_contagem[título] = 1
```

```python
títulos_ordenados = sorted(títulos_contagem.items(), key=lambda x: x[1], reverse=True)
```

```python
títulos, contagens = zip(*títulos_ordenados)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(títulos[:10], contagens[:10])
plt.xticks(rotation=45, ha="right")
plt.xlabel("Título")
plt.ylabel("Número de Visualizações")
plt.title("Títulos Mais Assistidos")
plt.tight_layout()
```
